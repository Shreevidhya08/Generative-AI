```python
from groq import Groq
```

```python
api_key = "your generated groq api key"
```

MODEL 1- llama-3.1-8b-instant

```python
client = Groq(api_key=api_key)

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Explain Artificial Intelligence in simple words."}
    ],
    temperature=0.6,
    max_tokens=100
)

print(response.choices[0].message.content)
```

MODEL 2 - openai/gpt-oss-120b

```python
client = Groq(api_key=api_key)

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": "You are an expert tutor."},
        {"role": "user", "content": "Describe the causes and effects of climate change."}
    ],
    temperature=0.6,
    max_tokens=200
)

# Extract and clean output
output = response.choices[0].message.content
output = output.strip()          # remove extra spaces/newlines

print("\n--- Model Output ---\n")
print(output)
print("\n--------------------")
```

MODEL 3- moonshotai/kimi-k2-instruct

```python
client = Groq(api_key=api_key)

response = client.chat.completions.create(
    model="moonshotai/kimi-k2-instruct",
    messages=[
        {"role": "user", "content": "List three major benefits of solar energy."}
    ],
    temperature=0.5,
    max_tokens=100
)

print("Kimi-K2 Response:", response.choices[0].message.content)
```
