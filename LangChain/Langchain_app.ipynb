```python
text = """
Artificial Intelligence enables machines to mimic human intelligence.
Machine Learning is a subset of AI that learns from data.
Deep Learning uses neural networks with many layers.
AI is widely used in healthcare, finance, and education.
"""

with open("sample.txt", "w") as f:
    f.write(text)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# 1. Load document
loader = TextLoader("sample.txt")
documents = loader.load()

# 2. Split into chunks (core LangChain feature)
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=80,
    chunk_overlap=10
)
docs = text_splitter.split_documents(documents)

# 3. Create embeddings (free, local)
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# 4. Store in vector database
vectorstore = FAISS.from_documents(docs, embeddings)

# 5. Retrieve relevant chunks
query = "Where is AI used?"
retrieved_docs = vectorstore.similarity_search(query, k=2)

# 6. Display results
for i, doc in enumerate(retrieved_docs, 1):
    print(f"\nResult {i}:")
    print(doc.page_content)
```
